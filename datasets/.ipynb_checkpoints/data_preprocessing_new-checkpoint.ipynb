```python
# Extract data from NetCDF file
def extract_data(nc_atm_fid, atm_fields, nc_hgt, nc_lnd_fid, lnd_fields, h_true = True, autoregress = True):
    lats = nc_atm_fid.variables['lat'][:]  # extract/copy the data
    lons = nc_atm_fid.variables['lon'][:]
    time = nc_atm_fid.variables['time'][:]
    field = []
    for f in atm_fields:
        field.append(np.array([resize(img, (64,64), preserve_range = True) for img in nc_atm_fid.variables[f][:]]))
        
    
    for j in lnd_fields:
        field.append(np.array([resize(img, (64,64), preserve_range = True) for img in nc_lnd_fid.variables[j][:]]))
    
    if autoregress:
        temp_swe = [resize(img, (64,64), preserve_range = True) for img in nc_lnd_fid.variables["SWE"][:]]
        field.append(np.array(temp_swe[len(temp_swe)-1:] + temp_swe[:len(temp_swe)-1]))
    
    if h_true:
        hgt = resize(((f_hgt.variables['topography'][:] >= 0) * f_hgt.variables['topography'][:])[0], (64, 64), preserve_range = True)
        temp = np.tile(hgt, (4018, 1, 1))
        field.append(temp)
    field = np.stack(field)
    return lats, lons, time, field
```

```python
atm_inp_var = [u'Prec', u'Tmax', u'Tmin', u'wind']
lnd_inp_var = [u'NetRad']
lats_in, lons_in, time_in, input_data = extract_data(f_atm, atm_inp_var, f_hgt, f_lnd, lnd_inp_var, True, False)
```
